### ğŸ’¡ Tips Penggunaan

### âš ï¸ Troubleshooting Umum

**1. Codespace loading lama**

- Tutup tab lain yang tidak perlu
- Refresh browser
- Coba akses di private/incognito mode

**2. File tidak terupload**

- Pastikan koneksi internet stabil
- Coba upload file satu per satu
- Gunakan file zip jika file banyak

**3. Jupyter Notebook tidak jalan**

- Install Jupyter: `pip install jupyter notebook`
- Restart kernel: Kernel â†’ Restart
- Pastikan Python kernel terpilih

**4. Package Python tidak ditemukan**

- Install di Terminal: `pip install package-name`
- Atau buat file `requirements.txt` dan jalankan: `pip install -r requirements.txt`
